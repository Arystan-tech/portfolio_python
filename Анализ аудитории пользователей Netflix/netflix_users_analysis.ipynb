# Приведение столбцов к стилю camel_case

```python
df = pd.DataFrame(df)

# Переименуем столбцы в стиль camelCase
df.rename(columns={
    'User ID': 'userId',
    'Subscription Type': 'subscriptionType',
    'Monthly Revenue': 'monthlyRevenue',
    'Join Date': 'joinDate',
    'Last Payment Date': 'lastPaymentDate',
    'Country': 'country',
    'Age': 'age',
    'Gender': 'gender',
    'Device': 'device',
    'Plan Duration': 'planDuration'
}, inplace=True)

# Проверим результат
print(df)
print(df.columns)
```

# Приведение типов данных

```python
#привести дату к типу datetime
# Приводим столбцы с датами к типу datetime
df['joinDate'] = pd.to_datetime(df['joinDate'], format='%d-%m-%y')
df['lastPaymentDate'] = pd.to_datetime(df['lastPaymentDate'], format='%d-%m-%y')

# Проверим результат
print(df)
print(df.dtypes)
```

```python
# изучть динамику привлечения пользователей (Join Date)

# Convert 'joinDate' to datetime format
df['join_date'] = pd.to_datetime(df['join_date'])

# Extract the month and year from joinDate
df['joinYearMonth'] = df['join_date'].dt.to_period('M')  # Group by month

# Group by the 'joinYearMonth' and count the number of users who joined each month
monthly_user_counts = df.groupby('joinYearMonth').size()

# Plot the user acquisition trend over time
plt.figure(figsize=(10, 6))
monthly_user_counts.plot(kind='line', marker='o', color='b')

# Add labels and title
plt.title('User Acquisition Over Time')
plt.xlabel('Join Date (Month)')
plt.ylabel('Number of Users Joined')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# изучть динамику выручки сервиса (накопленное Monthly Revenue по месяцу с момента Join Date до момента Last Payment Date)

# Convert 'joinDate' and 'lastPaymentDate' to datetime format
df['join_date'] = pd.to_datetime(df['join_date'])
df['lastPaymentDate'] = pd.to_datetime(df['lastPaymentDate'])

# Create a DataFrame to store monthly revenue by month
monthly_revenue_data = []

# Loop through each user to calculate revenue for each month they were active
for _, row in df.iterrows():
    start_date = row['join_date']
    end_date = row['lastPaymentDate']
    revenue_per_month = row['monthlyRevenue']

    # Loop through months from joinDate to lastPaymentDate
    while start_date <= end_date:
        monthly_revenue_data.append({
            'Month': start_date.to_period('M'),
            'Revenue': revenue_per_month
        })
        # Move to the next month
        start_date = (start_date + pd.DateOffset(months=1))

# Create a DataFrame from the monthly revenue data
monthly_revenue_df = pd.DataFrame(monthly_revenue_data)

# Group by month and calculate total revenue
monthly_revenue_summary = monthly_revenue_df.groupby('Month').sum()

# Plot the accumulated revenue over time
plt.figure(figsize=(10, 6))
monthly_revenue_summary.plot(kind='line', marker='o', color='green')

# Add labels and title
plt.title('Cumulative Monthly Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Accumulated Monthly Revenue')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# разбить клиентов на корзины по возрасту и определить самую платежеспособную группу

# Define the age bins and labels
age_bins = [18, 24, 34, 44, 54, 64, 100]  # Adjust the age bins as needed
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Create a new column 'age_group' that assigns each user to an age bin
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Group by the age group and calculate the total or average monthly revenue
age_group_revenue = df.groupby('age_group')['monthlyRevenue'].agg(['sum', 'mean'])

# Sort by total revenue to identify the most lucrative group
most_lucrative_group = age_group_revenue.sort_values(by='sum', ascending=False).head(1)

# Plot the total revenue for each age group
import matplotlib.pyplot as plt

# Plotting total revenue by age group
plt.figure(figsize=(10, 6))
age_group_revenue['sum'].plot(kind='bar', color='lightblue', edgecolor='black')

# Add labels and title
plt.title('Total Monthly Revenue by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Monthly Revenue')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()

# Display the most lucrative group
print("Most lucrative age group based on total revenue:")
print(most_lucrative_group)
```

```python
# посмотреть зависимость типа тарифа от группы возраста пользователя

# Define the age bins and labels
age_bins = [18, 24, 34, 44, 54, 64, 100]  # Adjust the age bins as needed
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']

# Create a new column 'age_group' that assigns each user to an age bin
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Group by age group and subscription type, then count the number of users in each group
age_subscription_counts = df.groupby(['age_group', 'subscriptionType']).size().unstack(fill_value=0)

# Plot the results as a stacked bar chart
ax = age_subscription_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=['lightblue', 'lightgreen', 'lightcoral'])

# Add labels and title
plt.title('Subscription Type Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# посмотреть зависимость типа тарифа от страны пользователя

# Group by country and subscription type, then count the number of users in each group
country_subscription_counts = df.groupby(['country', 'subscriptionType']).size().unstack(fill_value=0)

# Plot the results as a stacked bar chart
ax = country_subscription_counts.plot(kind='bar', stacked=True, figsize=(12, 8), color=['lightblue', 'lightgreen', 'lightcoral'])

# Add labels and title
plt.title('Subscription Type Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
