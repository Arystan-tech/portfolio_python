# Приведение столбцов к стилю camel_case

```python
def to_camel_case(column_name):
    words = column_name.split()
    return words[0].lower() + ''.join(word.title() for word in words[1:])

# Применение функции к названиям столбцов
df.columns = [to_camel_case(col) for col in df.columns]

# Вывод результата
print(df.head())
```

# Приведение типов данных

```python
#привести дату к типу datetime

df['orderDate'] = pd.to_datetime(df['orderDate'], format='%d-%m-%y')
```

# Поиск дублей и пропусков в данных

```python
df = df.drop_duplicates()
df = df.dropna()
```

# Исследовательский анализ данных

```python
# сколько уникальных заказов?

unique_orders = df['orderId'].nunique()
print(f"\nКоличество уникальных заказов: {unique_orders}")
```

```python
# сколько уникальных клиентов?

unique_customers = df['customerId'].nunique()
print(f"\nКоличество уникальных клиентов: {unique_customers}")
```

```python
# изучить локацию пользователей (визуализировать)

plt.figure(figsize=(12, 10))
sns.countplot(y='state', data=df, order=df['state'].value_counts().index)
plt.title('Распределение пользователей по штатам')
plt.xlabel('Количество пользователей')
plt.ylabel('Штат')
plt.xticks(rotation=70)
plt.show()
```

```python
# Изучение динамики количества заказов по дням
# Добавление столбца для месяцев (с извлечением месяца и года)
df['month'] = df['orderDate'].dt.to_period('M')

# Группировка по месяцам и подсчет количества заказов
orders_by_month = df.groupby('month')['orderId'].nunique().reset_index()
orders_by_month.rename(columns={'orderId': 'orderCount'}, inplace=True)

# Визуализация столбчатой диаграммы для количества заказов по месяцам
plt.figure(figsize=(12, 6))
sns.barplot(data=orders_by_month, x='month', y='orderCount', color='skyblue')

# Настройка заголовка и подписей осей
plt.title('Столбчатая диаграмма количества заказов по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Количество заказов')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()

# Создание полного списка всех дат от минимальной до максимальной
full_dates = pd.date_range(start=df['orderDate'].min(), end=df['orderDate'].max(), freq='D')

# Группировка по дате и подсчет количества заказов
orders_by_date = df.groupby(df['orderDate'].dt.date)['orderId'].nunique().reset_index()
orders_by_date.rename(columns={'orderDate': 'date', 'orderId': 'orderCount'}, inplace=True)

# Приведение столбца 'date' к типу datetime
orders_by_date['date'] = pd.to_datetime(orders_by_date['date'])

# Создание DataFrame с полными датами
orders_by_date_full = pd.DataFrame({'date': full_dates})

# Объединяем с полным списком всех дат, чтобы добавить пропущенные дни
orders_by_date_full = pd.merge(orders_by_date_full, orders_by_date, on='date', how='left')

# Заполнение пропусков в данных (если для некоторых дней нет заказов)
orders_by_date_full['orderCount'].fillna(0, inplace=True)

# Визуализация линейного графика для количества заказов по дням
plt.figure(figsize=(12, 6))
sns.lineplot(data=orders_by_date_full, x='date', y='orderCount', marker='o')

# Настройка заголовка и подписей осей
plt.title('Динамика количества заказов по дням')
plt.xlabel('Дата')
plt.ylabel('Количество заказов')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```

```python
# изучть динамику продаж по дате заказов

# Группировка по дате и подсчет суммы продаж
sales_by_date = df.groupby(df['orderDate'].dt.date)['sales'].sum().reset_index()
sales_by_date.rename(columns={'orderDate': 'date', 'sales': 'totalSales'}, inplace=True)

# Приведение столбца 'date' к типу datetime
sales_by_date['date'] = pd.to_datetime(sales_by_date['date'])

# Создание полного списка всех дат от минимальной до максимальной
full_dates = pd.date_range(start=df['orderDate'].min(), end=df['orderDate'].max(), freq='D')

# Создание DataFrame с полными датами
sales_by_date_full = pd.DataFrame({'date': full_dates})

# Объединяем с полным списком всех дат, чтобы добавить пропущенные дни
sales_by_date_full = pd.merge(sales_by_date_full, sales_by_date, on='date', how='left')

# Заполнение пропусков в данных (если для некоторых дней нет продаж)
sales_by_date_full['totalSales'].fillna(0, inplace=True)

# Визуализация динамики продаж по дням
plt.figure(figsize=(12, 6))
sns.lineplot(data=sales_by_date_full, x='date', y='totalSales', marker='o')

# Настройка заголовка и подписей осей
plt.title('Динамика продаж по дням')
plt.xlabel('Дата')
plt.ylabel('Сумма продаж')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```

```python
# изучть распределение продаж по категориям заказов

# Группировка по категории и подсчет суммы продаж
sales_by_category = df.groupby('category')['sales'].sum().reset_index()

# Визуализация распределения продаж по категориям с помощью столбчатой диаграммы
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_category, x='category', y='sales', color='skyblue')

# Настройка заголовка и подписей осей
plt.title('Распределение продаж по категориям')
plt.xlabel('Категория')
plt.ylabel('Сумма продаж')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```

```python
# изучть распределение продаж по городам (штатам)

# Группировка по штату и подсчет суммы продаж
sales_by_state = df.groupby('state')['sales'].sum().reset_index()

# Визуализация распределения продаж по штатам с помощью столбчатой диаграммы
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_state, x='state', y='sales', color='skyblue')

# Настройка заголовка и подписей осей
plt.title('Распределение продаж по штатам')
plt.xlabel('Штат')
plt.ylabel('Сумма продаж')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```

```python
# найти самые популярные товары (Product Name)

# Группировка по названию товара и подсчет суммы количества заказов
popular_products = df.groupby('productName')['quantity'].sum().reset_index()

# Сортировка товаров по количеству заказов в порядке убывания
popular_products = popular_products.sort_values(by='quantity', ascending=False)

# Вывод 10 самых популярных товаров
top_10_popular_products = popular_products.head(10)

# Вывод результата
print("Топ 10 самых популярных товаров:")
print(top_10_popular_products)
```

```python
# изучить распределение категорий внутри городов (штатам)

# Группировка по штату и категории товара, подсчет количества товаров в каждой категории
category_by_state = df.groupby(['state', 'category'])['productName'].count().reset_index()

# Визуализация распределения категорий товаров по штатам
plt.figure(figsize=(14, 8))
sns.barplot(data=category_by_state, x='state', y='productName', hue='category')

# Настройка заголовка и подписей осей
plt.title('Распределение категорий товаров по штатам')
plt.xlabel('Штат')
plt.ylabel('Количество товаров')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```

```python
# изучить динамику продаж внутри категорий

# Группировка по категории и дате, подсчет суммы продаж для каждой категории
sales_by_category_date = df.groupby([df['orderDate'].dt.date, 'category'])['sales'].sum().reset_index()

# Визуализация динамики продаж по категориям
plt.figure(figsize=(14, 8))
sns.lineplot(data=sales_by_category_date, x='orderDate', y='sales', hue='category', marker='o')

# Настройка заголовка и подписей осей
plt.title('Динамика продаж внутри категорий')
plt.xlabel('Дата')
plt.ylabel('Сумма продаж')

# Поворот меток оси X для лучшей читаемости
plt.xticks(rotation=45, ha='right', fontsize=10)

# Отображение сетки
plt.grid(True)

# Показать график
plt.show()
```
